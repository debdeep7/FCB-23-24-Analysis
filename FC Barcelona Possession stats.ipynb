```python
import pandas as pd
raw_df = pd.read_csv(r'C:\Users\anand\Desktop\FCB Stats\FC Barcelona Possession stats.csv')
raw_df
```

```python
#Removing the goalkeepers and the players with 90s less than 10
filtered_df = raw_df[raw_df['Pos'] != 'GK']
filtered_df.drop(filtered_df[filtered_df['90s'] < 10].index, inplace = True)
filtered_df.reset_index(inplace = True)
del filtered_df['index']
filtered_df
```

```python
#Importing the matplotlib libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Comparing the number of touches
plt.title('Amount of touches each player had throughout the entire season', loc = 'center', y = 1.5)
plt.pie(filtered_df['Touches'], labels = filtered_df['Player'], autopct = '%.1f%%', radius = 2) 
plt.show()
```

```python
#Separating the defenders
defenders_df = filtered_df[filtered_df['Pos'].str.startswith('DF')]
defenders_df
```

```python
#Comparing aggregate touches based on the different regions of the field
defenders_df['Aggregate Touches'] = defenders_df['Def Pen'] + defenders_df['Def 3rd'] + defenders_df['Mid 3rd'] + defenders_df['Att 3rd'] + defenders_df['Att Pen']
x = defenders_df['Aggregate Touches']
y = defenders_df['Player']
plt.xlabel('Aggregate Touches')
plt.ylabel('Players')
plt.barh(y,x, color = 'orange')
plt.show()
```

```python
#Successful take-ons % - The more the better
defenders_df['Successful take-ons %'] = defenders_df['Success']*100/defenders_df['Attempted Take-ons']
x = defenders_df['Successful take-ons %']
y = defenders_df['Player']
plt.xlabel('Take-ons success rate')
plt.ylabel('Players')
plt.barh(y,x, color = 'green')
plt.show()
```

```python
#% of take-ons failed - The less the better
defenders_df['% of take-ons failed'] = defenders_df['Tackled']*100/defenders_df['Attempted Take-ons']
x = defenders_df['% of take-ons failed']
y = defenders_df['Player']
plt.xlabel('% of take-ons failed')
plt.ylabel('Players')
plt.barh(y,x, color = 'cyan')
plt.show()
```

```python
#Percentage of progressive carrying distance - The more the better
defenders_df['Progressive carrying distance %'] = defenders_df['Progressive Carrying Dist']*100/defenders_df['Total Dist']
x = defenders_df['Progressive carrying distance %']
y = defenders_df['Player']
plt.xlabel('Progressive carrying distance %')
plt.ylabel('Players')
plt.barh(y,x, color = 'pink')
plt.show()
```

```python
#Attacking carries - The more the better
defenders_df['Attacking carries'] = defenders_df['Progressive Carries'] + defenders_df['Final Third'] + defenders_df['Penalty Area']
plt.pie(defenders_df['Attacking carries'], labels = defenders_df['Player'], autopct = '%.0f%%')
plt.title('Attacking Carries')
plt.show()
```

```python
#Visualization of miscontrols - The less the better
plt.pie(defenders_df['Miscontrols'], labels = defenders_df['Player'], autopct = '%.1f%%')
plt.title('Miscontrols')
plt.show()
```

```python
#Visualization of number of times dispossessed - The less the better
plt.pie(defenders_df['Dispossessed'], labels = defenders_df['Player'], explode = [0.1,0.1,0.1,0.1,0.2,0.1,0.1], autopct = '%.1f%%')
plt.title('Number of times dispossessed')
plt.show()
```

```python
#Separating the midfielders
midfielders_df = filtered_df[filtered_df['Pos'].str.startswith('MF')]
midfielders_df
```
