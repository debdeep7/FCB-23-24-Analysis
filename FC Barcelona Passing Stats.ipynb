```python
#Importing pandas libraries
import pandas as pd
#Importing the dataframe
raw_df = pd.read_csv(r'C:\Users\anand\Desktop\FCB Stats\FC Barcelona Passing stats.csv')
raw_df
```

```python
#Cleaning the dataframe
#Removing players with 90s less than 10
filtered_df = raw_df[raw_df['90s'] >= 10]
filtered_df.drop(filtered_df[filtered_df['Player'] == 'Squad Total'].index, inplace = True)
filtered_df.drop(filtered_df[filtered_df['Player'] == 'Opponent Total'].index, inplace = True)
filtered_df
```

```python
#Calculating various pass completion percentages
filtered_df['Total Pass Completion %'] = filtered_df['Total Passes Completed']*100/filtered_df['Total Passes Attempted']
filtered_df['Short Pass Completion %'] = filtered_df['Short Passes Completed']*100/filtered_df['Short Passes Attempted']
filtered_df['Medium Pass Completion %'] = filtered_df['Medium Passes Completed']*100/filtered_df['Medium Passes Attempted']
filtered_df['Long Pass Completion %'] = filtered_df['Long Passes Completed']*100/filtered_df['Long Passes Attempted']

final_filtered_df = filtered_df.round(2)
final_filtered_df
```

```python
#Separating the goalkeepers
goalkeepers_df = final_filtered_df[final_filtered_df['Pos'] == 'GK']
goalkeepers_df
```

```python
#Importing matplotlib libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Comparing the various pass completion percentages

#1 Total Pass Completion %
x = goalkeepers_df['Player']
y = goalkeepers_df['Total Pass Completion %']
plt.xlabel('Player')
plt.ylabel('%')
plt.title('Total Pass Completion %')
plt.bar(x,y, width = 0.2)
plt.show()

#2 Short Pass Completion %
x = goalkeepers_df['Player']
y = goalkeepers_df['Short Pass Completion %']
plt.xlabel('Player')
plt.ylabel('%')
plt.title('Short Pass Completion %')
plt.bar(x,y, width = 0.2, color = 'red')
plt.show()

#3 Medium Pass Completion %
x = goalkeepers_df['Player']
y = goalkeepers_df['Medium Pass Completion %']
plt.xlabel('Player')
plt.ylabel('%')
plt.title('Medium Pass Completion %')
plt.bar(x,y, width = 0.2, color = 'black')
plt.show()

#4 Long Pass Completion %
x = goalkeepers_df['Player']
y = goalkeepers_df['Long Pass Completion %']
plt.xlabel('Player')
plt.ylabel('%')
plt.title('Long Pass Completion %')
plt.bar(x,y, width = 0.2, color = 'orange')
plt.show()
```

```python
#Separating the defenders
defenders_df = final_filtered_df[final_filtered_df['Pos'].str.startswith('DF')]
defenders_df
```

```python
#Comparing the various pass completion percentages for defenders

#1 Total Pass Completion %
x = defenders_df['Player']
y = defenders_df['Total Pass Completion %']
plt.xlabel('%')
plt.ylabel('Player')
plt.title('Total Pass Completion %')
plt.barh(x,y, height = 0.5)
plt.grid()
plt.show()

#2 Short Pass Completion %
x = defenders_df['Player']
y = defenders_df['Short Pass Completion %']
plt.xlabel('%')
plt.ylabel('Player')
plt.title('Short Pass Completion %')
plt.barh(x,y, height = 0.5, color = 'red')
plt.grid()
plt.show()

#3 Medium Pass Completion %
x = defenders_df['Player']
y = defenders_df['Medium Pass Completion %']
plt.xlabel('%')
plt.ylabel('Player')
plt.title('Medium Pass Completion %')
plt.barh(x,y, height = 0.5, color = 'black')
plt.grid()
plt.show()

#4 Long Pass Completion %
x = defenders_df['Player']
y = defenders_df['Long Pass Completion %']
plt.xlabel('%')
plt.ylabel('Player')
plt.title('Long Pass Completion %')
plt.barh(x,y, height = 0.5, color = 'orange')
plt.grid()
plt.show()
```

```python
#Assist analysis
def assess_performance(row):
    if row['Assists'] > row['xA']:
        return 'Overperformed'
    elif row['Assists'] < row['xA']:
        return 'Underperformed'
    else:
        return 'Performed as expected'

defenders_df['xA Performance'] = defenders_df.apply(assess_performance, axis=1)
defenders_df
```
