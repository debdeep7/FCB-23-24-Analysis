```python
import pandas as pd
df = pd.read_csv(r'C:\Users\anand\Desktop\FCB Stats\FC Barcelona Defensive Actions stats.csv')
df
```

```python
df1 = df.drop(df[df['Pos'] == 'GK'].index)
df2 = df1.drop(df[df['Player'] == 'Squad Total'].index)
df3 = df2.drop(df[df['Player'] == 'Opponent Total'].index)
df3
```

```python
df4 = df3[df3['Pos'].str.contains('DF')]
df4
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df4['Tackle Success Rate'] = df4['Tackles Won']/df4['Tackles']
x = df4['Player']
y = df4['Tackle Success Rate']
plt.xlabel('Tackle Success Rate')
plt.ylabel('Players')
plt.barh(x,y, height = 0.5)
```

```python
df5 = df3[df3['Pos'].str.contains('MF')]
df5
```

```python
#removing the players with less than 15 tackles
df6 = df5[df5['Tackles'] >= 15]
df6
```

```python
df6['Tackle Success Rate'] = df6['Tackles Won']/df6['Tackles']
x = df6['Player']
y = df6['Tackle Success Rate']
plt.xlabel('Tackle Success Rate')
plt.ylabel('Players')
plt.barh(x,y, height = 0.5)
```

```python
import numpy as np
x = df4['Shots Blocked']
y = df4['Passes Blocked']
z = df4['Player']
xpos = np.arange(len(z))
plt.yticks(xpos,z)
plt.xlabel('Blocks')
plt.ylabel('Players')
plt.barh(xpos - 0.2,x, label = 'Shots Blocked', height = 0.5)
plt.barh(xpos + 0.2,y, label = 'Passes Blocked', height = 0.5)
plt.legend()
```
