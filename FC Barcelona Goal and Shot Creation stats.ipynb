```python
#Importing the pandas libraries and our data
import pandas as pd
raw_df = pd.read_csv(r'C:\Users\anand\Desktop\FCB Stats\FC Barcelona Goal and Shot Creation stats.csv')
raw_df
```

```python
#Cleaning the df by removing players having 90s less than 10
filtered_df = raw_df[raw_df['90s'] >= 10]
filtered_df.drop(filtered_df[filtered_df['Player'] == 'Squad Total'].index, inplace = True)
filtered_df.drop(filtered_df[filtered_df['Player'] == 'Opponent Total'].index, inplace = True)
filtered_df
```

```python
#Removing the players who have 0 creative actions
filtered_df.drop(filtered_df[filtered_df['Shot Creating Actions'] == 0].index, inplace = True)
filtered_df.drop(filtered_df[filtered_df['Goal Creating Actions'] == 0].index, inplace = True)
filtered_df
```

```python
#Importing matplotlib libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Visualising the shot creating actions
plt.pie(filtered_df['Shot Creating Actions'], labels = filtered_df['Player'])
plt.show()
```

```python
#Visualising the goal creating actions
plt.pie(filtered_df['Goal Creating Actions'], labels = filtered_df['Player'])
plt.show()
```

```python
#Measuring the quality of creative action - Calculating the fraction of goal creating actions upon shot creating actions - The more the better
filtered_df['Quality of Areative Action'] = filtered_df['Goal Creating Actions']/filtered_df['Shot Creating Actions']
filtered_df
```
